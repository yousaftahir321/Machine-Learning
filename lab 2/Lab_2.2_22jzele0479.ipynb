## Pandas Data Structure
Panda has two types of Data Structure:<br> 
    a) Series<br>
    b) Dataframe
### Series
It is a one dimentional array with indexes. It stores a single column or row of data in a Dataframe.<br>
It is a one dimensional array capable of holding any type of data.
### Dataframe
It is a tabular spreadsheet like structure representing rows each of which contains one or more columns.<br>
It is a multi dimensional array capable of holding any type of data.<br>
__Series__ is like a column, a __DataFrame__ is the whole table.

## Slicing through loc function

```python
import pandas as pd

# Sample DataFrame
data = {
    "Name": ["Ali", "Sara", "Bilal", "Hina"],
    "Marks": [85, 92, 76, 89],
    "Age": [20, 21, 19, 22]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
# 1. Select row with label 2
print(df.loc[2])
```

```python
# 2. Select specific rows and columns by NAME
print(df.loc[1:3, ["Name", "Marks"]])
```

```python

# 3. Applying Filter: Conditional selection: students with Marks > 85
print(df.loc[df["Marks"] > 85])
```

##  Slicing through iloc

```python
df
```

```python
# 1. Select row at index 0
print(df.iloc[0])
```

```python
# 2. Select 2 rows and 2 columns by position
print(df.iloc[0:2, 0:2])
```

```python
# 3. Select last row using negative index
print(df.iloc[-1])
```

# Groupby
The groupby method allows you to group rows of data together and call aggregate functions

​
​

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

### Now you can use the .groupby() method to group rows together based off of a column name.
For instance let's group based off of Company. This will create a DataFrameGroupBy object:

```python
df.groupby('Company')
```

##### You can save this object as a new variable:

```python
by_comp = df.groupby("Company")
type(by_comp)
```

#### And then call aggregate methods of the object:

```python
by_comp.std(numeric_only=True)
```

```python
df.groupby('Company').std(numeric_only=True)
```

```python
by_comp.std(numeric_only=True)
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```
